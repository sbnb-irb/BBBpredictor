### Checking the results of the prototype predictor

```python
import pandas as pd
```

```python
trueVal= pd.read_csv("../DATA/BBBP.csv")
trueVal.set_index('num',inplace=True)
```

```python
predictions= pd.read_csv("../singularity/test_file_2_BBBpredictions.tsv", sep='\t',header=[1,2])
predictions.index.rename('num', inplace=True)
predictions.columns= ['pred']
```

```python
predictions.head()
```

```python
alltog = pd.concat([trueVal,predictions],axis=1)
alltog
```

```python
alltog[['p_np','pred']]
```

```python
# Numbers of signatures that cannot be calculated
len(alltog[alltog.pred==-1])
```

```python
len(alltog[alltog.p_np != alltog.pred]) # 50 wrong predictions or cannot calculate signature
```

```python
# Percentage of wrong predictions
len(alltog[alltog.p_np != alltog.pred]) *100 / len(alltog)
```
